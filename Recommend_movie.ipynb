```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
credits = pd.read_csv('credits.csv') 
```

```python
movie = movies.merge(credits, on = "title")
```

```python

movie.info()
```

```python
movie = movie[["movie_id", "keywords" ,"title" , "cast" , "crew" , "overview", "genres"]]
```

```python
movie.head(5)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
        
```

```python
movie.dropna(inplace=True)
```

```python
movie.isnull().sum()
```

```python
movie.head()
```

```python
movie["keywords"] = movie["keywords"].apply(convert)
movie["genres"] = movie["genres"].apply(convert)
```

```python
movie.head()
```

```python
import ast
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter += 1   
    return L
```

```python
movie["cast"] = movie["cast"].apply(convert3)
```

```python
movie.head()
```

```python
movie['cast'] = movie['cast'].apply(lambda x:x[0:3])
```

```python
movie.head(5)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie.iloc[1571]
```

```python
movie.head()
```

```python
def replace(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
        
    return L1
```

```python
movie["cast"] = movie["cast"].apply(replace)
movie["crew"] = movie["crew"].apply(replace)
movie["keywords"] = movie["keywords"].apply(replace)
movie["genres"] = movie["genres"].apply(replace)
```

```python
movie.head()
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
movie["tags"] = movie["overview"] + movie["genres"] + movie["keywords"] + movie["cast"] + movie["crew"]
```

```python
movie.head()
```

```python
movies = movie.drop(columns = ["overview", "genres" , "keywords","cast", "crew"])
```

```python
movies.head()
```

```python
movies["tags"] = movies["tags"].apply( lambda x: " ".join(x))
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vector = cv.fit_transform(movies["tags"]).toarray()
```

```python
vector
```

```python
print(cv)
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
movies[movies['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movies_re):
    index = movies[movies["title"] == movies_re].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
